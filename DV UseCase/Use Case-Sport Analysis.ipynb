```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ipl_players_dataset.csv')
df = df.dropna()
df = df.drop_duplicates()
df.head()
```

```python
fig = px.scatter(df,
                 x="StrikeRate",
                 y="Runs",
                 color="Player",
                 size="Matches",
                 title="Player Performance Analysis")
fig.show()
```

```python
players = ["Virat Kohli", "MS Dhoni", "Rohit Sharma"]
df_filtered = df[df["Player"].isin(players)]

fig = px.bar(df_filtered,
             x="Player",
             y="Runs",
             color="Player")
fig.show()
```

```python
plt.figure(figsize=(8,6))
corr = df[['Runs','StrikeRate','Matches','Average']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python

fig = px.area(
    df,
    x="Season",
    y="Runs",
    color="Player",
    template="plotly_dark",
    title="Runs Trend Across Seasons"
)

fig.show()
```

```python
fig = px.parallel_coordinates(
    df,
    color="Runs",
    dimensions=["Runs","StrikeRate","Matches","Average"],
    color_continuous_scale=px.colors.diverging.Tealrose,
    template="plotly_dark",  
    title="Multi-Dimensional Player Analysis"
)

fig.show()
```

```python
fig = px.box(
    df,
    x="Player",
    y="Runs",
    color="Player",
    template="plotly_dark",   
    title="Player Performance Distribution"
)

fig.show()
```
