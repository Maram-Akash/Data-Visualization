```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# -------------------------------
# SAMPLE DATA (you can replace with CSV file)
# -------------------------------
data = {
    "Date": pd.date_range(start="2024-01-01", periods=30, freq="D"),
    "Temperature": [30, 32, 31, 33, 35, 36, 34, 32, 31, 30,
                    29, 28, 30, 31, 33, 35, 36, 37, 38, 39,
                    37, 36, 34, 33, 32, 31, 30, 29, 28, 27],
    "Humidity": [70, 72, 68, 65, 60, 58, 62, 64, 66, 68,
                 70, 72, 74, 75, 76, 78, 80, 82, 81, 79,
                 77, 75, 73, 71, 70, 69, 68, 67, 66, 65],
    "Pressure": [1012, 1013, 1011, 1010, 1009, 1008, 1007, 1006, 1005, 1004,
                 1003, 1002, 1001, 1000, 1002, 1004, 1006, 1008, 1010, 1012,
                 1013, 1014, 1015, 1013, 1012, 1011, 1010, 1009, 1008, 1007],
    "Rain": [0, 1, 0, 0, 2, 3, 1, 0, 0, 0,
             0, 1, 2, 3, 0, 0, 1, 2, 0, 0,
             1, 0, 0, 2, 3, 1, 0, 0, 0, 1]
}

df = pd.DataFrame(data)

# Convert date
df['Date'] = pd.to_datetime(df['Date'])

# -------------------------------
# 1. Temperature Over Time (Line Graph)
# -------------------------------
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Temperature'], marker='o', color='blue')
plt.title("Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.grid()
plt.show()

# -------------------------------
# 2. Temperature Trend Line (Regression)
# -------------------------------
X = np.arange(len(df)).reshape(-1, 1)
y = df['Temperature'].values

model = LinearRegression()
model.fit(X, y)
trend = model.predict(X)

plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Temperature'], label="Actual Temperature", marker='o')
plt.plot(df['Date'], trend, label="Trend Line", color='red')
plt.title("Temperature Trend Line")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# -------------------------------
# 3. Humidity Area Chart
# -------------------------------
plt.figure(figsize=(10,5))
plt.fill_between(df['Date'], df['Humidity'], color='skyblue', alpha=0.5)
plt.plot(df['Date'], df['Humidity'], color='blue')
plt.title("Humidity Area Chart")
plt.xlabel("Date")
plt.ylabel("Humidity (%)")
plt.xticks(rotation=45)
plt.grid()
plt.show()

# -------------------------------
# 4. Weather Parameters Over Time
# (Temperature, Humidity, Pressure)
# -------------------------------
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Temperature'], label="Temperature")
plt.plot(df['Date'], df['Humidity'], label="Humidity")
plt.plot(df['Date'], df['Pressure'], label="Pressure")
plt.title("Weather Parameters Over Time")
plt.xlabel("Date")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
