```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (replace with your file path)
# Error: FileNotFoundError - The file 'diabetes.csv' was not found.
# Please ensure 'diabetes.csv' is uploaded to your Colab environment
# or provide the correct path to the file (e.g., '/content/diabetes.csv').
diabetes = pd.read_csv("diabetes.csv")

# --- Categorical Data ---
# Bar chart
sns.countplot(x="Outcome", data=diabetes)
plt.title("Bar Chart of Diabetes Outcome")
plt.show()

# Pie chart
diabetes['Outcome'].value_counts().plot.pie(autopct='%1.1f%%', labels=['No Diabetes','Diabetes'])
plt.title("Pie Chart of Diabetes Outcome")
plt.show()

# --- Continuous Data ---
# Scatter plot
plt.scatter(range(len(diabetes)), diabetes['Glucose'])
plt.title("Scatter Plot of Glucose Levels")
plt.xlabel("Patient Index")
plt.ylabel("Glucose")
plt.show()

# Line plot
plt.plot(diabetes['Glucose'])
plt.title("Line Plot of Glucose Levels")
plt.xlabel("Patient Index")
plt.ylabel("Glucose")
plt.show()

# Strip plot
sns.stripplot(x="Outcome", y="Glucose", data=diabetes)
plt.title("Strip Plot of Glucose by Outcome")
plt.show()

# Swarm plot
sns.swarmplot(x="Outcome", y="Glucose", data=diabetes)
plt.title("Swarm Plot of Glucose by Outcome")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
employee = pd.read_csv("/content/Employee.csv")

# --- Categorical Data ---
# Bar chart
sns.countplot(x="Education", data=employee)
plt.title("Bar Chart of Employees by Education")
plt.show()

# Pie chart
employee['Education'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart of Employees by Education")
plt.show()

# --- Continuous Data ---
# Histogram
sns.histplot(employee['Age'], bins=20, kde=False)
plt.title("Histogram of Employee Age")
plt.show()

# Density plot
sns.kdeplot(employee['Age'], shade=True)
plt.title("Density Plot of Employee Age")
plt.show()

# Rug plot
sns.kdeplot(employee['Age'], shade=True)
sns.rugplot(employee['Age'])
plt.title("Density + Rug Plot of Employee Age")
plt.show()
```
