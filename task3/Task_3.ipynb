1.Categorical vs categorical

Bar chart


```python
import matplotlib.pyplot as plt
ct = pd.crosstab(df['sex'], df['address'])

ct.plot(kind='bar', stacked=True)
plt.title("Stacked Bar Chart")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

Grouped Bar Chart

```python
ct.plot(kind='bar')
plt.title("Grouped Bar Chart")
color=['grey','black']
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

Segmented Bar Chart

```python
ct_percent = ct.div(ct.sum(axis=1), axis=0)

ct_percent.plot(kind='bar', stacked=True)
plt.title("Segmented Bar Chart (100%)")
plt.xlabel("Sex")
plt.ylabel("Proportion")
plt.show()
```

Mosaic Plot

```python
from statsmodels.graphics.mosaicplot import mosaic

mosaic(df, ['sex', 'address'])
plt.title("Mosaic Plot")
plt.show()
```

CONTINUOUS vs CONTINUOUS

Scatterplot with Fit Line


```python
import seaborn as sns
sns.lmplot(x='studytime', y='G3', data=df)
plt.title("Scatterplot with Fit Line")
plt.show()
```

CATEGORICAL vs CONTINUOUS

Bar Chart

```python
import seaborn as sns
sns.barplot(x='sex', y='G3', data=df)
plt.title("Bar Chart (Mean G3 by Sex)")
plt.show()
```

Grouped Kernel Density Plot

```python
import seaborn as sns
sns.kdeplot(data=df, x='G3', hue='sex', fill=True)
plt.title("Grouped Kernel Density Plot")
plt.show()
```

Box Plot

```python
import seaborn as sns
sns.boxplot(x='sex', y='G3', data=df)
plt.title("Box Plot")
plt.show()
```

 Violin Plot

```python
import seaborn as sns
sns.violinplot(x='sex', y='G3', data=df)
plt.title("Violin Plot")
plt.show()
```
